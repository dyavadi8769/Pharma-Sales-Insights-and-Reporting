```python
# Analyse Dimentions
categorical_cols = pharma_df.select_dtypes(include='object').columns
for col in pharma_df[categorical_cols]:
    print("****** "+ col+ " ******:")
    print(pharma_df[col].value_counts())
    print()
    print()
```

```python
for col in pharma_df.select_dtypes(include=['object']):
    print(f"{col}: {pharma_df[col].nunique()} unique values")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

numeric_cols = pharma_df.select_dtypes(include=['int64','float64']).columns
for col in numeric_cols:
    plt.figure()
    pharma_df[col].hist(bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
corr = pharma_df[numeric_cols].corr()
plt.figure(figsize=(8,8))
plt.imshow(corr, interpolation='nearest', aspect='auto')
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=90)
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()
```
